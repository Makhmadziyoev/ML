```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DfAli2.csv')
df
```

```python
import datetime as DT
for i, strdate in enumerate(df['Date']):
    dt = DT.datetime.strptime(strdate, '%Y-%m-%d %H:%M:%S')
    dt2 = dt.timestamp()
    df['Date'][i] = dt2
    print(i, dt2)
```

```python
for i, znach in enumerate(df['Source']):
    if(znach == 'THK' ):
        znach2 = 111
        df['Source'][i] = znach2
    elif(znach == 'TIBANNE_LIMITED_HK'):
        znach3 = 333
        df['Source'][i] = znach3
    else:
        df['Source'][i] = int(znach)
    print(i, type(df['Source'][i]))    
```

```python
from sklearn.model_selection import train_test_split
```

```python
points_train, points_test, labels_train, labels_test = train_test_split(df.iloc[:, :-1], df['label'], test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python

ss = StandardScaler()
ss.fit(points_train)
points_train.iloc[:, :] = ss.transform(points_train)
points_test.iloc[:, :] = ss.transform(points_test)
```

```python
points_new.iloc[:, :] = ss.transform(points_new)
```

```python
print(points_train.shape, points_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(points_train, labels_train)
```

```python
prediction = knn.predict(points_test)
print(prediction)
print(points_test.assign(predict=prediction))
```

```python
print(format(knn.score(points_test, labels_test)))
```
